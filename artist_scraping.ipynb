```python
file = open('artist.txt','r')
myData = [["pkID", "mbID", "Artist"]]
for line in file:
    x = file.readline().split('\t')
    row = [x[0], x[1], x[2]]
    myData.append(row)

import csv 
myFile = open('artist.csv', 'w')
with myFile:
    writer = csv.writer(myFile)
    writer.writerows(myData)
print("Writing complete")    
```

## 3. refine '/mbdump-derived/mbdump/tag'
we get '../Cleared Data/tag.txt', which links tag ID with tagname


```python
#Author: Humberto Hernandez
#Last Updated: 5/31/2018 5:14pm
fnew = open('../Cleared Data/tag.txt', 'w')
headers = 'tag ID\ttagname\n'
fnew.write(headers)
 
with open(path + '/mbdump-derived/mbdump/tag') as f:
    for line in f:
        line_list = line.split('\t')
        new_line_list = line_list[0:2]
        delimiter = '\t'
        new_line = delimiter.join(new_line_list)
        fnew.write(new_line)
        fnew.write('\n')

fnew.close()    
```

## 5. use '../Cleared Data/final_artist_table.csv' to create MBID dictionary
we get '../Cleared Data/MBID_dictionary.p' that stores the dictionary

```python
MBID_indices_dict = artist_and_tag_table.groupby('MBID').groups
MBID_dictionary = {}
for key in MBID_indices_dict:
    index_list = MBID_indices_dict[key]
    tag_list= []
    for index in index_list:
        tag = artist_and_tag_table.loc[index,'tagname']
        if not(pd.isnull(tag)):
            tag_list.append(tag)
    if len(tag_list) > 0:
        MBID_dictionary[key] = tag_list
```
